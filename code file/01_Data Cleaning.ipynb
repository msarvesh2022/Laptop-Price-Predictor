## Data cleaning

```python
df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['HD_display'] = df.ScreenResolution.apply(lambda x: 1 if 'HD' in x else 0)
```

```python
df.head()
```

```python
df['Cpu_name'] = df.Cpu.apply(lambda x: " ".join(x.split()[0:3]))
```

```python
def find(processor):
    if processor == "Intel Core i7" or processor == "Intel Core i5" or processor == "Intel Core i3":
        return processor
    else:
        if processor.split()[0]== "Intel":
            return "Other Intel processor"
        else:
            return "AMD"
```

```python
df['cpu_brand'] = df.Cpu_name.apply(find)
```

```python
df.head()
```

```python
df['GH'] = df.Cpu.apply(lambda x: " ".join(x.split()[4:5]))
```

```python
df.head()
```

```python
df.GH.isnull().sum()
```

```python
def check(processor):
    if processor == "Intel Core i7" or processor == "Intel Core i5" or processor == "Intel Core i3":
        return processor
    else:
        if processor.split()[0]== "Intel":
            return "Other Intel processor"
        else:
            return "AMD"
```

```python
df['cpu_brand'] = df.Cpu_name.apply(check)
```

```python
df.head()
```

```python
df.drop(columns=['Cpu','Cpu_name'], inplace=True)
```

```python
df['Ram'] = df['Ram'].str.replace('GB',"")
```

```python
df.Weight= df.Weight.str.replace("kg","")
```

```python
df.head()
```

```python
df['memory'] = df['Memory'].apply(lambda x: " ".join(x.split()[0:1]))
```

```python
df.memory= df.memory.str.replace("GB","")
```

```python
df.memory= df.memory.str.replace("TB","000")
```

```python
df.head()
```

```python
df['gpu_brand'] = df.Gpu.apply(lambda x: x.split()[0])
```

```python
df.head()
```

```python
df.OpSys.value_counts()
```

```python
def find(os):
    if os == "Windows 10" or os == "Windows 10 S" or os == "Windows 7" or os =='NO OS' :
        return "Windows"
    elif os=='macOS' or 'Mac OS X':
        return 'MacOS'
    elif os=='Linux':
        return 'Linux'
    else :
        return 'Other Operating Sysytem'
        
```

```python
df['operating_system'] = df.OpSys.apply(find)
```

```python
df.operating_system.value_counts()
```

```python
df.head()
```

```python
df.isnull().mean()
```

```python
df= df.drop(columns=['ScreenResolution','Memory','Gpu', 'OpSys'], axis =1 )
```
