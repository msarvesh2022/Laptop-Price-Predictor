```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=  pd.read_csv('clean_laptop_dataset.csv')
```

```python
df
```

```python
X= df.iloc[:,0:10]
y= df.iloc[:,-1]
```

```python
num_data= X.select_dtypes(exclude=['object'])
cat_data= X.select_dtypes(include=['object'])
```

```python
num_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
scaler.fit(num_data)
```

```python
X_num= scaler.transform(num_data)
```

```python
X_num = pd.DataFrame(X_num, columns=num_data.columns)
```

```python
X_num
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit(cat_data)
```

```python
encoded = ohe.transform(cat_data)
```

```python
ohe_columns= ohe.get_feature_names_out(cat_data.columns)
```

```python
ohe_columns
```

```python
encoded = encoded.toarray()
```

```python
cat_dataframe= pd.DataFrame(encoded,columns=ohe_columns)
```

```python
X=X_num.merge(cat_dataframe,left_index=True, right_index=True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2 )
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr= LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred= lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, f1_score, mean_absolute_error
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```
